# Comparison of the two following SMOTE index selection implementations:
### Rand
```
samples_indices = random_state.randint(
    low=0, high=nn_num.size, size=n_samples)
rows = np.floor_divide(samples_indices, nn_num.shape[1])
cols = np.mod(samples_indices, nn_num.shape[1])
```

### Tile
```
min_cls_idxs = np.arange(X.shape[0])
reps = math.ceil(n_samples / len(min_cls_idxs))
rows = np.tile(min_cls_idxs, reps)[:n_samples]
cols = random_state.choice(self.k_neighbors, size=len(rows))
```
